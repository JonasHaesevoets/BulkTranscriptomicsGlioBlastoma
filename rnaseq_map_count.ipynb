For my assignment I will review the RNA-seq of a control glioblastoma celline and a glioblastoma celline who has been treated with NVP-BEZ235. The glioblastoma celline was chosen in this research due to the fact that this cancer is the most lethal primary brain cancer without effective molcular targeted therapies. Glioblastomas are characterised by activation of the PI3K/AKT/mTOR pathway, the tested compound is a dual specific inhibitor to this pathway by inhibiting both PI3K and mTOR while also inhibiting genes involved in glycolysis. For my assignment I will investigate the effect of NVP-BEZ235 treatment and its impact on the metabolism in glioblastoma cells.






control genes: c-myc is a potent oncogene that is frequently overexpressed in many human cancers

               LDHA is involved in ATP-production from aerobic glycolysis and is downregulated in response to NVP-BEZ235
               
               ENO1 is a glycolotic gene that is upregulated in glioblastoma and downregulated in response to NVP-BEZ235
               
               EGFR is a receptor for epidermal growth factor and is overexpressed in glioblastoma
               
               FOXO3 is a biomarker to assess efficicacy of the PI3K/AKT inhibition by NVP-BEZ235
